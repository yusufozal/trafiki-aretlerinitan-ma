```python
cam =cv2.VideoCapture(0)
face_xml = cv2.CasecadeClassifier('/content/haarcascade_frontalface_default.xml')

while(cam.isOpened()):
    ret,frame=cam.read()
    
    frame_gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    my_result=face_xml.detecMultiscale(frame_gray,1.3,3)

    for x,y,w,h in my_result:
        pt1=(x,y)
        pt2=(x+w,y+h)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
        cv2.putText(img=frame,texet="yusuf",org=(10,60),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScLe=(1),color=(255,0,0))
       
    cv2.imshow("frame",frame)

    if cv2.waitKey(1) & 0xFF==ord("q"):
        break


cam.release()
cv2.destroyAllWindows

```
